```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12,8)
```

## Bestimmung der Erdbeschleunigung $g$

```python
h = np.array([0,2,4,6,8,10])
t = np.array([0,0.62,0.91,1.12,1.20,1.44])
t_sq = t**2
```

```python
t_theo = np.linspace(0, 1.5, 100)
h_theo = 0.5*9.81*t_theo**2

fig, ax = plt.subplots()
fig.suptitle('h gegen t', fontsize=16)
#ax.set_yscale('log')
ax.scatter(t,h, label='Messwerte')
ax.plot(t_theo, h_theo, label='theoretischer Verlauf', linestyle='--')
ax.legend()
#ax.plot(x, ye)
ax.grid(which='both', linestyle='--')
ax.set_xlabel('t [s]')
ax.set_ylabel('h [m]')

```

```python
t_sq_theo = np.linspace(0,2.2, 100)
h_theo = 0.5*9.81*t_sq_theo

fig, ax = plt.subplots()
fig.suptitle('h gegen $t^2$', fontsize=16)

ax.scatter(t_sq,h, s=50, label='Messwerte')
#-----------hier ausblenden--------------#
#ax.plot(t_sq_theo, h_theo, linestyle='--', label='theoretischer Verlauf')
#-----------hier ausblenden--------------#
ax.legend()

ax.grid(which='both', linestyle='--')
ax.set_xlabel('$t^2$ [s]')
ax.set_ylabel('h [m]')
```
