```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = np.array([0., 0.01, 0.02, 0.03, 0.04])
I = [1, 0.55, 0.22, 0.14, 0.07]
log_I = np.log(I)
x = np.linspace(-0.01, 0.05, 100)
a, b = np.polyfit(d, log_I, 1)
def poly(x, a, b): 
    return a*x + b
y = poly(x, a, b)

ye = np.exp(a*x+b)
```

```python
fig, ax = plt.subplots()
ax.scatter(d, log_I, label='Messdaten')
#ax.set_yscale('log')
ax.plot(x, y)
ax.legend()
#ax.plot(x, ye)
ax.grid(which='both', linestyle='--')
ax.set_xlabel('d [m]')
ax.set_ylabel('I(d) [$W/m^2$]')
ax.set_xlim(-0.005, 0.045)
```

```python
fig, ax = plt.subplots()
ax.scatter(d, I, label='Messdaten')
#ax.set_yscale('log')
#ax.plot(x, y)
ax.legend()
ax.plot(x, ye)

ax.grid(which='both', linestyle='--')
ax.set_xlabel('d [m]')
ax.set_ylabel('I(d) [$W/m^2$]')
ax.set_xlim(-0.005, 0.045)
```
